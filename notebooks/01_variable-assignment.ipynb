## References in Action: Variables

There's another word we can start adding to our vocabularies when it comes to references which are _variables_.  A variable is a basic space in memory that we can store data in that we can then _refer_ to in the future.

> **Do you already know other programming langauges?**
>
> Python has the concept of **global** and **private** scopes when it comes to variable references.  Private variables are encapsulated inside of functions and lambda context and global can be defined and accessed in the _main_ context as you can see so far.  Also the **global** keyword can be used in methods and functions when you want to differentiate between private and global when the namespace is the same.  For example `global alpha` will change the context of `alpha` in a function to be global to all references after the statement.

To illustrate the power of variables in programming, lets consider the following code examples.

````python
## The cost of rent of a 3 bedroom, 2 bath home in San Francisco
3 * 1400 + 2 * 124
````

````python
## The cost of rent of a 2 bedroom, 1 bath home in San Francisco
2 * 1400 + 1 * 124
````

Later on, we will be talking about linear regression and how it can be used to estimate many problems that have linear relationships between data features and the thing to be estimated.  To simplify this idea, say I came up with an inaccurate but somewhat "in the ballpark" assumption about how to estimate the price of rents in San Francisco like this:
    
## $(\text{bedrooms}*\text{some number}) + (\text{bathrooms}*\text{some number}) = \text{rent of SF apartment}$

We could start a business estimating rent prices (you can be CEO).  Let's try some calculations!

> **Did you know**
>
> You can use comments with the `#` sign anywhere inside of your Python code?  It makes your code easier to read and will help you remember what it was you were working on prior to looking at your code again in the future.

````python
print(1 * 1400 + 1 * 124)  # 1 bedroom apartment!
print(6 * 1400 + 2 * 124)  # 6 bedrooms apartment!
print(20 * 1400 + 2 * 124) # 20 bedrooms apartment!
print(1 * 1400 + 50 * 124) # 1 bed 50 bathroom apartment *very* practical when having guests!
````

### It works!  But wait, it's a little hard to read everything!?

Yeah it is.  If you wanted to test a bunch bedrooms and bathroom combinations to see how well they estimate you will have to look at each line individually and mentally account for which number is for bedrooms, how much you're scaling bedrooms, bathrooms, how much you're scaling bathrooms to get a rent price estimate.

### We could use variables!

What if we put the numbers we scale bedrooms and bathrooms into variables? We will take the same code as previously defined and make a small adjustment to the code.  Can you tell what it is?  Read each line and try to figure out what what's different vs the last example.

````python
bedroom_scaler = 1400
bathroom_scaler = 124

print(1 * bedroom_scaler + 1 * bathroom_scaler)  # 1 bedroom apartment!
print(6 * bedroom_scaler + 2 * bathroom_scaler)  # 6 bedrooms apartment!
print(20 * bedroom_scaler + 2 * bathroom_scaler) # 20 bedrooms apartment!
print(1 * bedroom_scaler + 50 * bathroom_scaler) # 1 bed 50 bathroom apartment *very* practical when having guests!
````

- We added `bedroom_scaler = 1400` on line 1
- We added `bathroom_scaler = 124` on line 2

Then we replaced every instance from the original example with the cooresponding variables.  Rather than typing "1400" for every bedroom, we can use type `bedroom_scaler` instead.  Same for `bathroom_scaler`.

### Other problems we solve

- If we wanted to build a model for the Boston and Miami markets based on the same assumption, we only need to change the first two lines of code and viola, we don't have to repeat repetive work.

- If you name your variables descriptively, it's easy for other humans to read.

- Your code is easier to scan positionally so if you want to differentiate between which values are for bedrooms and which are for batrhooms, it's a lot easier to tell.

### Third rule of Python Variable club: No reserved words allowed!

Within Python, there are keywords that exist which control the behavior of Python code and are actually fundamental to how it works.  So we can't use those words!  

Here's a list of the words (case-sensitive):
```
False     def       if        raise
None      del       import    return
True      elif      in        try
and       else      is        while
as        except    lambda    with
assert    finally   nonlocal  yield
break     for       not
class     from      or
continue  global    pass
```

It's time to break some more rules.  Go forth and assign a variable with the same name as one of the reserved words!

> **Reserved words** in Python are special words that control the behavior of Python in different contexts.  As engineers, we have to be careful of not using these in variable scope so that Python can differentiate between the instructions you give it and the variables you assign.  For instance if you were allowed to use `if = "not in my town partner"`, then it would be hard for Python to differentiate when you wanted to write an `if` statement.  Fortunately, Python does a good job of enforcing this standard.  Sometimes it's easy to forget that these words aren't allowed as variable names but if you're seeing a syntax error in your code but everything looks right otherwise, it's possible you've used a **reserved word** as the variable name.
